```python
import os
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/home/kakade/Downloads/sk_corrected_shruti_SZ .csv')
```

```python
data.columns
```

```python
data['date'] = pd.to_datetime(data['date'], format="%d/%m/%y")
data = data.set_index('date').sort_index()
```

```python
f = plt.figure(figsize=(12,8))
plt.plot(data['total_cum'], marker='o', label='total')
plt.plot(data['total_confirmed_cum'], marker='o', label='total_confirmed')
plt.plot(data['total_suspected_cum'], marker='o', label='total_suspected')
plt.plot(data['isolated_cum'], marker='o', label='isolated')
plt.plot(data['suspected_under_testing_cum'], marker='o', label='suspected_under_testing')
plt.plot(data['suspected_negative_cum'], marker='o', label='suspected_negative')
plt.plot(data['discharged_cum'], marker='o', label='discharged')
plt.plot(data['deceased_cum'], marker='o', label='deceased')
plt.legend()
plt.yscale('log')
plt.title('Cumulative count of original columns')
plt.grid(True)
plt.show()
f.savefig('/home/kakade/Cumulative count of original columns.png')
```
